# 12 : SPLIT DATAFRAME INTO 2 SUBSETS

```python
import seaborn as sns 
import pandas as pd 

df = sns.load_dataset('titanic')
df.head()
```

```python
# LEN AND SHAPE FUNCTION IS USED TO GET THE NUMBER OF ROWS AND COLUMNS

len(df)
df.shape
```

```python
from random import random

# HALF OF THE DATA FROM DATAFRAME STORE INTO KHASTI_1

khasti_1 = df.sample(frac=0.50, random_state=1)
khasti_1.shape
```

```python
# FIRST HALF DROP FROM THE MAIN DATAFRAME AND ANOTHER DATA STORE INTO KHASTI_2 

khasti_2 = df.drop(khasti_1.index)
khasti_2.shape

# IN THAT WAY WE SPLIT DATAFRAME INTO TWO SUBSETS
```

# 13 : JOIN TWO DATASETS

```python
frames = [khasti_1, khasti_2]
df_1 = pd.concat(frames)
df_1.shape
```
