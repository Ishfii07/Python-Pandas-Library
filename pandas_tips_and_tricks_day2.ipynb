# 06 : REVERSE ROW ORDER 

```python
import seaborn as sns 
import pandas as pd 

df = sns.load_dataset('titanic')
df.head()
```

```python
# GETTING ROWS INTO REVERSE ORDER USING THIS SYNTAX

df.loc[::-1].head()

# [start:stop:step]

# START : Kaunsi index se slicing start karni hai. Agar empty chhorda jai to default start hota hai.
# STOP : Kis index par slicing end hogi. Agar empty chhorda jai to end tak hoti hai.
# STEP: Kitne steps ka difference hoga. Positive step se forward jaoge aur negative step se backward

# [::-1] ka Matlab:

# start or stop empty hain, iska matlab list ya DataFrame ke start se end tak dekho.
# step (-1) hai, iska matlab sequence ko reverse kar do.
```

```python
# AB MAI CHATA HO KE ORDER REVERSE HE RAHE LAKIN INDEX JO HAI NA WO 0 SE START HO JAYE BAKI ORDER REVERSE HE RAHE

df.loc[::-1].reset_index(drop=True).head()
```

# 07 : REVERSE COLUMN ORDER 

```python
# REVERSE THE COLUMN ORDER 

df.loc[:, ::-1].head()

# [:, ::-1] KA MATLAB 

# ':' (ROW SELECTION) : ka matlab hai ke sabhi rows ko select karo. Ye start aur stop ke values nahi specify 
#                       karta, isliye ye default ke roop mein use hota hai, jo ke poore DataFrame ke rows ko 
#                       select karta hai.

# '::-1' (COLUMN SELECTION) : ka matlab hai columns ko reverse order mein select karna.
```

# 09 : CONVERT STRINGS TO NUMBERS 

```python
df = pd.DataFrame({
    'Col_A' : ['11', '21', '31', '41', '51'],
    'Col_B' : ['12', '22', '32', '42', '52'],
    })

df
```

```python
df.dtypes
```

```python
# CONVERT DATA TYPES TO ANY OTHER DATATYPES USING ASTYPE FUNCTION 

df.astype({'Col_A' : 'float64' , 'Col_B' : 'int64'}).dtypes
```

```python
pd.to_numeric(df['Col_A'], errors='coerce')

# errors='coerce' :  set kiya jata hai, to agar kisi value ko numeric type mein convert nahi kiya ja sakta 
#                    (jaise non-numeric text), to wo value NaN (Not a Number) se replace ho jati hai.

# errors='raise' :   Default behavior; agar koi conversion error hota hai to exception raise hota hai.

# errors='ignore':   Agar conversion error hota hai to original data waisa hi rehta hai (no conversion).
```

# 10 : REDUCE DATAFRAME SIZE : 

```python
df = sns.load_dataset('titanic')

# SHAPE FUCNTION GIVE THE TOTAL NUMBER OF ROWS AND COLUMNS IN THE DATAFRAME
df.shape
```

```python
# WHEN U GET THE RANDOM SAMPLE DATA FORM UR WHOLE DATAFRAME U USE THIS AND IT RETURN EVERY EXECUATION RANDOM ROWS OF THE DATAFRAME 
df.sample(frac=0.1).shape
```

```python
sample_df = df.sample(frac=0.1)
sample_df

# BUT YOU WANT ONLY FIXED RANDOM ROWS FROM DATAFRAME U SHOULD NOT NEED OF EVERY EXECUTION DIFFERENT DATA SO U CAN USE RANDOM_STATE= 42 (U PUT ANY INTEGER VALUE U WANT)

sample_df = df.sample(frac=0.1, random_state=20)
sample_df
```
