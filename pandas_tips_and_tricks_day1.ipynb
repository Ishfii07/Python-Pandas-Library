# 03 : HOW TO RENAME THE COLUMN NAME 

```python
df_1 = pd.DataFrame({'Col_1' : [11, 22, 33, 44, 55], 'Col_2' : [21, 22, 23, 24, 25]}) # no of items in list are same
df_1
```

```python
df_1.rename(columns={'Col_1' : 'A_Col', 'Col_2' : 'B_Col'}, inplace=True)
df_1
```

```python
# ANOTHER METHOD OF THE REPLACING THE NAME OF THE COLUMN
df_1.columns=['Col_AA', 'Col_BB']
df_1
```

```python
# TO REPLACE ANY CHARACTER AND STRING IN THE COLUMN

df_1.columns = df_1.columns.str.replace('_', '*')
df_1
```

```python
# ADD PREFIX TO THE COLUMN

df_1 = df_1.add_prefix('1_')
df_1
```

```python
# ADD SUFFIX TO THE COLUMN

df_1 = df_1.add_suffix('_0')
df_1
```

```python
df_1.columns = ['Col_A', 'Col_B']
df_1
```
